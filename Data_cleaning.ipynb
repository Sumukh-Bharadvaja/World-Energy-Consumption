```python
import pandas as pd
import numpy as np
```

Importing the data into dataframe

```python
data= pd.read_csv('World Energy Consumption.csv')
```

Ovewview of the data

```python
data
```

The data consists of around many countries data aggregated into 22012 rows with most of the entries to be Nan.

Let us see the number of values for each countries and Unions 

```python
data['country'].value_counts() 
```

As it can be seen, the data seems to have alot of abnormalities and redundencies. Lets use filter in order to reduce the number of columns to aquire only important which is required for displaying the energy prodcution


```python
df0 = data[['iso_code', 
           'country', 
           'year',
           'biofuel_electricity', 
           'hydro_electricity',
           'nuclear_electricity',
           'solar_electricity',
           'wind_electricity', 
           'other_renewable_electricity',
           'coal_electricity',
           'gas_electricity',
           'oil_electricity',
           'biofuel_elec_per_capita', 
           'hydro_elec_per_capita',
           'nuclear_elec_per_capita',
           'solar_elec_per_capita',
           'wind_elec_per_capita', 
           'coal_elec_per_capita',
           'gas_elec_per_capita',
           'oil_elec_per_capita',
           'population'
          ]]

df0
```

Merging different fuels' columns into one column called 'fuel' we use melt from pandas in order to do this

```python
df = df0.melt(id_vars=['iso_code', 'country', 'year', 'population'],
              var_name='fuel',
              value_name='production'
)

df
```

Let us take on of the cases 'Brazil' in order to use it as a control for transformation and cheking the aggreagtion

```python
df.loc[(df['country'] == 'Brazil') & (df['year'] == 2018)]
```

Creating the column 'per_capita' to move the values of fuel_per_capita in order to have separate column for it

```python
def split_type(row):
    if row['fuel'].find('per_capita') != -1:
        value = row['production']
    else:
        value = 0
    return value

df['per_capita'] = df.apply(split_type, axis=1)

df
```

Once again taking our use case to verfiy the aggregation

```python
df.loc[(df['country'] == 'Brazil') & (df['year'] == 2018)]
```

Now cleaning the values in column production that were moved to column per_capita

```python
def clean_production(row):
    if row['per_capita'] != 0:
        value = 0
    else:
        value = row['production']
    return value


df['production'] = df.apply(clean_production, axis=1)

df
```

Repeating the check with our use case

```python
df.loc[(df['country'] == 'Brazil') & (df['year'] == 2018)]
```

Cleaning the fuel names for clear picture

```python
df['fuel'] = df['fuel'].str.replace('_elec_per_capita', '')
df['fuel'] = df['fuel'].str.replace('_electricity', '')

df['fuel'].unique()
```

This is just an extra step I realized I need to do to make sure my 'Country' filter would work as intended in the dashboard.

```python
df = df.drop(df[df['country'] == 'World'].index)

df['country'].unique()
```

Finally, grouping the dataframe and checking the result with our use case

```python
df_final = df.groupby(['iso_code', 'country', 'year', 'population', 'fuel']).sum().reset_index()

df_final.loc[(df_final['country'] == 'Brazil') & (df_final['year'] == 2018)]
```
